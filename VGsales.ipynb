```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style(style = 'darkgrid')
```

```python
games = pd.read_csv('vgsales.csv', index_col= 'Rank')
```

```python
games.head()
```

```python
games.info()
```

```python
games.columns
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(games.isnull(), cbar = False, yticklabels= False, cmap = 'prism')
```

```python
print('Countplot of the platforms. The top 2 are DS and PS2')
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
sns.countplot(x = 'Platform', data = games, palette = 'prism')
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
sns.countplot(x = 'Genre', data = games, palette = 'gist_rainbow')
print('Action games has the most sales')
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
sns.countplot(x = 'Year', data = games.dropna(), palette = 'gist_ncar')
print('2018 and 2019 had the most sales')
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
plt.yticks(rotation='horizontal')
sns.heatmap(games.dropna().corr(), annot = True, yticklabels= False, cmap = 'rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.jointplot('NA_Sales', 'EU_Sales', data = games, kind= 'reg', color = 'black')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'Genre' , y = 'Global_Sales', data = games, palette='magma', color='maroon')
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x = 'Genre' , y = 'Other_Sales', data = games, palette='magma', color='maroon')
```
